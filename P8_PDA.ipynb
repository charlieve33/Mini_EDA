```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tampilkan visualisasi dalam notebook
%matplotlib inline

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/SEM 4/Kecerdasan Buatan/Praktikum P8/Dataset/data_balita.csv')

# Menampilkan 20 data teratas
print("=== 20 Data Teratas ===")
print(df.head(20))

# Menampilkan 20 data terbawah
print("\n=== 20 Data Terbawah ===")
print(df.tail(20))

# Menjelaskan tipe data setiap kolom
print("\n=== Tipe Data Setiap Kolom ===")
print(df.dtypes)

# Menampilkan statistik deskriptif
print("\n=== Statistik Deskriptif ===")
print(df.describe(include='all'))

# Mengecek missing value
print("\n=== Cek Missing Value ===")
print(df.isnull().sum())

# Menampilkan distribusi data
print("\n=== Distribusi Data (Histogram Setiap Kolom Numerik) ===")
df.hist(figsize=(15,10), bins=20)
plt.tight_layout()
plt.show()

# Menghitung korelasi minimal 3 pasang kolom numerik
print("\n=== Korelasi Tiga Pasang Kolom Numerik ===")
numerik = df.select_dtypes(include=np.number)
corr_matrix = numerik.corr()
print(corr_matrix)

# Menampilkan korelasi tiga pasang kolom
pairs = list(corr_matrix.columns[:3])
for i in range(len(pairs)):
    for j in range(i+1, len(pairs)):
        print(f"Korelasi antara {pairs[i]} dan {pairs[j]}: {corr_matrix.loc[pairs[i], pairs[j]]}")

# Visualisasi: Histogram
plt.figure(figsize=(8,5))
sns.histplot(data=df, x=pairs[0], kde=True)
plt.title(f'Histogram dari {pairs[0]}')
plt.show()

# Visualisasi: Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(data=df[pairs])
plt.title(f'Boxplot dari {pairs}')
plt.show()
```
